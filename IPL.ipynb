## Convert Date Column

```python
matches['date'] = pd.to_datetime(matches['date'])
```

```python
matches['date'].dtype
```

## Handle Missing Values

```python
import warnings
```

```python
## city Column (51 nulls)

matches['city'] = matches['city'].fillna('Unknown')
```

```python
matches['city'].fillna('Unknown', inplace=True)
```

```python
## player_of_match (5 nulls)

matches['player_of_match'].fillna('No Award', inplace=True)
```

```python
## winner (5 nulls)

matches['winner'].fillna('No Result', inplace=True)
```

```python
## result_margin (19 nulls) 

matches['result_margin'].fillna(0, inplace=True)
```

```python
## target_runs & target_overs

matches['target_runs'].fillna(0, inplace=True)
```

```python
matches['target_overs'].fillna(0, inplace=True)
```

## method Column

- method → DLS method matches only

```python
matches['method'].fillna('Normal', inplace=True)
```

## Standardize Team Names

This is VERY IMPORTANT for:

- SQL joins
- Power BI
- Accurate analysis

```python
team_replacements = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Kings XI Punjab': 'Punjab Kings',
    'Deccan Chargers': 'Sunrisers Hyderabad',
    'Rising Pune Supergiant': 'Rising Pune Supergiants',
    'Royal Challengers Bengaluru':'Royal Challengers Bangalore'
}

columns_to_replace = [
    'team1',
    'team2',
    'winner',
    'toss_winner'
]

for col in columns_to_replace:
    matches[col] = matches[col].replace(team_replacements)
```

```python
matches.duplicated().sum()
```

```python
matches.isnull().sum()
```

## PART B — CLEANING deliveries.csv

```python
## Standardize Team Names

for col in ['batting_team', 'bowling_team']:
    deliveries[col] = deliveries[col].replace(team_replacements)
```

```python
## Fill extras_type Nulls
deliveries['extras_type'].fillna('No Extra', inplace=True)
```

```python
## Fill dismissal columns
deliveries['player_dismissed'].fillna('Not Out', inplace=True)

```

```python
deliveries['dismissal_kind'].fillna('No Wicket', inplace=True)

```

```python
deliveries['fielder'].fillna('No Fielder', inplace=True)
```

```python
##Check Duplicates
deliveries.duplicated().sum()
```

```python
## Final Null Check
deliveries.isnull().sum()
```

```python
## Save Cleaned matches.csv
matches.to_csv("matches_cleaned.csv", index=False)
```

```python
##Save Cleaned deliveries.csv
deliveries.to_csv("deliveries_cleaned.csv", index=False)
```

## TOP BOWLERS ANALYSIS

```python
plt.figure(figsize=(12,6))

sns.barplot(
    x=top_bowlers.values,
    y=top_bowlers.index
)

plt.title("Top Wicket Takers")

plt.xlabel("Wickets")

plt.ylabel("Bowler")

plt.show()
```

## POWERPLAY ANALYSIS

```python
## Overs 1–6
powerplay = deliveries[
    deliveries['over'] <= 5
]

powerplay_runs = powerplay.groupby(
    'batting_team'
)['total_runs'].sum()

powerplay_runs = powerplay_runs.sort_values(
    ascending=False
)

powerplay_runs
```

## Insights



- Aggressive teams dominate powerplay.
- Teams with strong openers score faster early.

## DEATH OVER ANALYSIS

```python
## Overs 16–20
death_overs = deliveries[
    deliveries['over'] >= 15
]

death_runs = death_overs.groupby(
    'batting_team'
)['total_runs'].sum()

death_runs = death_runs.sort_values(
    ascending=False
)

death_runs
```

## Insights



- Mumbai Indians excel in finishing.
- Strong death hitters improve chase success.

## ADVANCED INSIGHTS

```python
##Strike Rate Analysis
balls = deliveries.groupby(
    'batter'
)['ball'].count()

runs = deliveries.groupby(
    'batter'
)['batsman_runs'].sum()

strike_rate = (
    runs / balls
) * 100

strike_rate = strike_rate.sort_values(
    ascending=False
)

strike_rate.head(10)
```

```python
#Economy Rate Analysis
runs_given = deliveries.groupby(
    'bowler'
)['total_runs'].sum()

balls_bowled = deliveries.groupby(
    'bowler'
)['ball'].count()

economy = (
    runs_given / balls_bowled
) * 6

economy = economy.sort_values()

economy.head(10)
```

## Strike Rate vs Total Runs

```python
runs = deliveries.groupby(
    'batter'
)['batsman_runs'].sum()

balls = deliveries.groupby(
    'batter'
)['ball'].count()

strike_rate = (runs / balls) * 100

batting_stats = pd.DataFrame({
    'Runs': runs,
    'StrikeRate': strike_rate
})

batting_stats = batting_stats[
    batting_stats['Runs'] > 2000
]

plt.figure(figsize=(12,6))

sns.scatterplot(
    x='Runs',
    y='StrikeRate',
    data=batting_stats
)

plt.title("Strike Rate vs Total Runs")

plt.show()
```

## Insight

- Identifies aggressive AND consistent batsmen.
